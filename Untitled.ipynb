```python
cursor.execute("SELECT brand_name, COUNT(*) FROM Smartphones GROUP BY brand_name")
data = cursor.fetchall()

brands = [row[0] for row in data]
counts = [row[1] for row in data]

plt.figure(figsize=(10, 6))
plt.bar(brands, counts, color="royalblue")
plt.title("ბრენდების მიხედვით სმარტფონების რაოდენობა")
plt.xlabel("ბრენდი")
plt.ylabel("რაოდენობა")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# ამ დიაგრამაში ჩანს რამდენი სმარტფონია ბრენდების მიხედვით
```

```python
cursor.execute("SELECT has_5g, COUNT(*) FROM Smartphones GROUP BY has_5g")
data_5g = cursor.fetchall()

labels = [row[0] for row in data_5g]
values = [row[1] for row in data_5g]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct="%1.1f%%", colors=["lightgreen", "salmon"])
plt.title("5G მხარდაჭერის მქონე სმარტფონების პროცენტული განაწილება")
plt.tight_layout()
plt.show()

# ამ pie chart-ში ჩანს თუ რამდენ სმარტფონს აქვს და არ აქვს 5G მხარდაჭერა
# true ნიშნავს რომ აქვს false ნიშნავს რომ არ აქვს ხოლო პატარაზე სადაც წერია 0.1% მონაცემი არაა
```

```python
cursor.execute("SELECT battery_capacity, price FROM Smartphones")
data = cursor.fetchall()

battery = []
price = []

for b, p in data:
    try:
        b_val = float(b)
        p_val = float(p)
        battery.append(b_val)
        price.append(p_val)
    except:
        continue

plt.figure(figsize=(8, 6))
plt.scatter(battery, price, color='purple', alpha=0.6)
plt.title("ბატარეის მოცულობა და ფასი")
plt.xlabel("ბატარეის მოცულობა (mAh)")
plt.ylabel("ფასი")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# ეს დიაგრამა აჩვენებს ბატარეის მოცულობასა და სმარტფონის ფასს შორის არსებობს თუ არა კავშირი
```
